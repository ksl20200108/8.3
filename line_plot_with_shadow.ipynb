```python
import seaborn as sns
import numpy as np
import traceback
xs = np.random.randint(low=1, high=101, size=20000)
ys = np.random.rand(20000)
different_things = ['a', 'b']
hues = np.random.choice(different_things, 20000, p=[0.5, 0.5])
print(ys)
```

```python
# import traceback
# try:
#   //
# except Exception:
#     traceback.print_exc()
```

### Confidence Interval

```python
line_plot1 = sns.lineplot(x=xs, y=ys, hue=hues).figure
line_plot1.savefig("../rl1.png")
```

### Standard Deviation

```python
line_plot2 = sns.lineplot(x=xs, y=ys, hue=hues, ci="sd").figure
line_plot2.savefig("../rl2.png")
```
